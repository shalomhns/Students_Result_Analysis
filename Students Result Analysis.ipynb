## Adding grades to each of the scores in the dataframe

#### We are creating new column for the grade of each students in the dataframe for all the subject offered by the students using the if statement

```python
# adding grade for each scores in the dataframe using if statement
def grading(x):
    if x <= 40:
        return "F"
    elif x <= 50:
        return "P"
    elif x <= 60:
        return "C"
    elif x <= 70:
        return "B"
    elif x<= 80:
        return "A"
    elif x<= 100:
        return "A*"
    else:
        return " "
```

```python
# Creating a new column for each subject grades within the Dataframe using the apply method
df_new["igbo"] = df_new['IGBO '].apply(grading)
df_new["bus std"] = df_new['BUS STD'].apply(grading)
df_new["cca"] = df_new['CCA'].apply(grading)
df_new["yor"] = df_new['YOR'].apply(grading)
df_new["crs"] = df_new['CRS'].apply(grading)
df_new["eng"] = df_new['ENG'].apply(grading)
df_new["bst"] = df_new['BST'].apply(grading)
df_new["pvs"] = df_new['PVS'].apply(grading)
df_new["nve"] = df_new['NVE'].apply(grading)
df_new["his"] = df_new['HIS'].apply(grading)
df_new["maths"] = df_new['MATHS'].apply(grading)
df_new["fre"] = df_new['FRE'].apply(grading)
```

## Calculating the percentage number of students that is above credit level

### It is the percentage of the total number of students offering the subject divide by the number of students at credit level and above

```python
IGBO = (df_creditlv['IGBO ']/df_total['IGBO '])*100
BUS_STD= (df_creditlv['BUS STD']/df_total['BUS STD'])*100
CCA = (df_creditlv['CCA']/df_total['CCA'])*100
YOR = (df_creditlv['YOR']/df_total['YOR'])*100
CRS = (df_creditlv['CRS']/df_total['CRS'])*100
ENG = (df_creditlv['ENG']/df_total['ENG'])*100
BST = (df_creditlv['BST']/df_total['BST'])*100
PVS = (df_creditlv['PVS']/df_total['PVS'])*100
NVE = (df_creditlv['NVE']/df_total['NVE'])*100
HIS = (df_creditlv['HIS']/df_total['HIS'])*100
MATHS = (df_creditlv['MATHS']/df_total['MATHS'])*100
FRE = (df_creditlv['FRE']/df_total['FRE'])*100
```

```python
# Laoding the values into a DataFrame 
df_percent = pd.DataFrame([{'IGBO':IGBO, 
                            'BUS_STD':BUS_STD, 
                            'CCA':CCA, 'YOR': YOR, 
                            'CRS':CRS,'ENG':ENG, 
                            'BST':BST, 'PVS':PVS, 
                            'NVE':NVE, 'HIS':HIS, 
                            'MATHS':MATHS, 'FRE':FRE}])
```
